```rust
:dep hidapi = { version = "2.6", default-features = false, features = ["linux-static-hidraw"] }

use hidapi;

pub const VENDOR_ID: u16 = 0x041e;
pub const PRODUCT_ID: u16 = 0x3256;
pub const INTERFACE: i32 = 4;

let api = hidapi::HidApi::new().unwrap();
let connection = api.open_path(std::ffi::CStr::from_bytes_with_nul(b"/dev/hidraw13\0").unwrap()).unwrap();

let mut cmd = [0u8; 65];
cmd[0] = 0x00;
cmd[1] = 0x5a;
cmd[2] = 0x11;
cmd[3] = 0x03;
cmd[4] = 0x01;
cmd[5] = 0x96;
cmd[6] = 0x07;

connection.write(&cmd).unwrap();

let mut buf = [0u8; 64];
loop {
    match connection.read_timeout(&mut buf, 500) {
        Ok(0) => { println!("-- timeout --"); break; }
        Ok(n) => { println!("Read {} bytes: {:02x?}", n, &buf[..n]); }
        Err(e) => { println!("Error: {:?}", e); break; }
    }
}
```
